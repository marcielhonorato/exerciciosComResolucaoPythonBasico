Você trabalha em uma empresa que tem 18 lojas espalhadas por todo o Brasil e divididas em 5 estados diferentes:
- RJ
- SP
- MG
- GO
- AM

Todo trimestre, são calculados os indicadores de cada funcionário de cada loja e esses indicadores são armazenados em um arquivo em Excel.

Cada estado tem 1 Gerente Geral responsável por todas as lojas daqueles estados.

Pediram para você enviar para cada Gerente Geral todas as bases de indicadores correspondentes às lojas que ele é responsável, porque a equipe deles precisa desses indicadores.

Obs: Não vamos enviar por e-mail porque ainda não aprendemos a fazer isso, mas vamos deixar todos os arquivos em uma pasta única para cada gerente, ou seja, para cada estado.

Então o seu desafio é separar todos os arquivos de forma que cada arquivo esteja na pasta do estado correspondente aquele arquivo.

Obs: Para pegar o nome de um arquivo como um texto no pathlib, você pode usar Path.name ou arquivo.name
caminho = Path('Pasta/Arquivo.csv')

print(caminho.name) -> resposta: 'Arquivo.csv'"

```python
# Resolução 1


from pathlib import Path
import shutil

caminho = Path('/home/marciel/Documents/exerciciosComResolucaoPythonBasico/base_dados/Arquivos_Lojas')
arquivos = caminho.iterdir()

origem = Path('/base_dados/Arquivos_Lojas')

dirRJ = Path('/home/marciel/Documents/exerciciosComResolucaoPythonBasico/base_dados/Lojas Brasil/Loja Rio de Janeiro')
dirSP = Path('/home/marciel/Documents/exerciciosComResolucaoPythonBasico/base_dados/Lojas Brasil/São Paulo')
dirMG = Path('/home/marciel/Documents/exerciciosComResolucaoPythonBasico/base_dados/Lojas Brasil/Minas Gerais')
dirGO = Path('/home/marciel/Documents/exerciciosComResolucaoPythonBasico/base_dados/Lojas Brasil/Goiania')
dirAM = Path('/home/marciel/Documents/exerciciosComResolucaoPythonBasico/base_dados/Lojas Brasil/Amazonas')

for arquivo in arquivos:
    if '_RJ' in str(arquivo):
        novo_caminho = dirRJ / arquivo.name
        print(novo_caminho)
        shutil.move(origem / arquivo, novo_caminho)
    elif '_SP' in str(arquivo):
        novo_caminho = dirSP / arquivo.name
        print(novo_caminho)
        shutil.move(origem / arquivo, novo_caminho)
    elif '_MG' in str(arquivo):
        novo_caminho = dirMG / arquivo.name
        print(novo_caminho)
        shutil.move(origem / arquivo, novo_caminho)
    elif '_GO' in str(arquivo):
        novo_caminho = dirGO / arquivo.name
        print(novo_caminho)
        shutil.move(origem / arquivo, novo_caminho)
    else:
        novo_caminho = dirAM / arquivo.name
        print(novo_caminho)
        shutil.move(origem / arquivo, novo_caminho)
```

```python
# Resolução 2

from pathlib import Path
import shutil

pastas = ['RJ','MG','SP','GO','AM']
for loja in pastas:
    Path('base_dados/Brasil/{}'.format(loja)).mkdir()
```

```python
caminho = Path('base_dados/Arquivos_Lojas')

arquivos = caminho.iterdir()

for arquivo in arquivos:
    nome_arquivo = arquivo.name
    if nome_arquivo[-3:] == 'csv':
        nome_estado = nome_arquivo[-6:-4]
        destino = Path('base_dados/Brasil') / Path('{}/{}'.format(nome_estado,nome_arquivo))
        shutil.move(arquivo, destino)
```
